```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("diabetes.csv")
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
```

```python
df.head()
```

```python
# first we need input and output
x = df.iloc[:,:-1]
y = df["Outcome"]
```

```python
x.head()
```

```python
y.head()
```

```python
df.shape
```

```python
lr = LogisticRegression()
# sfs = SequentialFeatureSelector(lr , k_features = 8, forward = True)
sfs = SequentialFeatureSelector(lr , k_features = 8, forward = False)
```

```python
sfs.fit(x , y)
```

```python
sfs.feature_names
```

```python
sfs.k_feature_names_
```

```python
sfs.k_score_
# this score tells how accurate output accuracy on given n number of features 
# k_score_ varies with number of selected features or columns
```
