```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("cgpa_iq.csv")
```

```python
df.drop(columns=["Student_ID"] , inplace = True)
```

```python
sns.kdeplot(data = df["IQ"])
plt.show()
```

# both are normal distribution manner so we can apply Gaussian Naive Bayes algo to predict ans

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x= df.iloc[:,:-1]
y = df["Placement"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = .20 , random_state = 42)
gnb = GaussianNB()
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_test , y_test)*100
```

```python
mnb = MultinomialNB()
mnb.fit(x_train , y_train)
```

# it has the best efficiency

```python
mnb.score(x_test , y_test)*100
```

```python
bnb = BernoulliNB()
bnb.fit(x_train , y_train)
```

```python
bnb.score(x_test , y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf =gnb )
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf =mnb )
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf =bnb )
plt.show()
```
