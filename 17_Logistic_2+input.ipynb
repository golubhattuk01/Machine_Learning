```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("cgpa_iq.csv")
```

```python
df.head()
```

```python
# grapah between cgpa and placement to  know they are linearly separable
plt.figure(figsize=(4 ,3))
sns.scatterplot(x= "CGPA" , y="Placement" ,data = df)
plt.show()
```

```python
# grapah between iq and placement to  know they are linearly separable
plt.figure(figsize=(4 ,3))
sns.scatterplot(x= "IQ" , y="Placement" ,data = df)
plt.show()
```

```python
# grapah between iq and cgpa to  know they are linearly separable
plt.figure(figsize=(4 ,3))
sns.scatterplot(x= "CGPA" , y="IQ" ,data = df , hue = "Placement")
plt.show()
```

```python
x = df.iloc[: , 1:3]
y = df["Placement"]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = .20 , random_state = 42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.score(x_test , y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = lr)
plt.show()
```
